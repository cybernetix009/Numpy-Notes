```python
#Splitting
```

```python
c = np.array([[1,2,3], [4,5,6]])
```

```python
print(c)
```

```python
split = np.hsplit(c,3)  #horizontal Split

for s in split:
    print(s)
```

```python
#vertical split
```

```python
vsplit = np.vsplit(c,2)

for s in vsplit:
    print(s)
```
